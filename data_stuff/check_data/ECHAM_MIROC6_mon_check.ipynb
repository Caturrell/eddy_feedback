```python
import xarray as xr
import os

import functions.data_wrangling as dw
```

# ECHAM6.3 (AWI-CM-1-1-MR)

## variable: epfy

```python
echam_epfy_path = os.path.join(path, 'epfy', 'ECHAM6.3_AWI', '*.nc')

echam_epfy = xr.open_mfdataset(echam_epfy_path, combine='nested', concat_dim='ens_ax')
echam_epfy = echam_epfy.epfy.sel(lon=0)
echam_epfy = dw.seasonal_mean(echam_epfy, season='djf', take_mean=True)
echam_epfy = echam_epfy.where(echam_epfy.level < 1000).load()
echam_epfy
```

```python
echam_epfy.plot(x='lat', y='level', col='ens_ax', col_wrap=10)
```

## variable: ua

```python
echam_ua_path = os.path.join(path, 'ua', 'ECHAM6.3_AWI', '*.nc')

echam_ua = xr.open_mfdataset(echam_ua_path, combine='nested', concat_dim='ens_ax')
echam_ua = echam_ua.ua.mean('lon').load()
echam_ua = dw.seasonal_mean(echam_ua, season='djf', take_mean=True)
echam_ua
```

```python
echam_ua.plot(x='lat', y='level', col='ens_ax', col_wrap=10)
```
