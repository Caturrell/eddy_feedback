```python
import pandas as pd
import numpy as np
import os
```

## New function (Regridded) vs Old function

This is a comparison between `calculate_efp()` and `calculate_efp_pamip()`. If I want to investigate in the future, it might be due to having the surface in the calculation...?
- Using regridded data

```python
data_path = '/home/links/ct715/eddy_feedback/reproduce/data/efp'
new_file = os.path.join(data_path, 'efp_NH_SH_mon_new-fn_regridded_3x3.csv') # see pamip_efp_maths_20250422
old_file = os.path.join(data_path, 'efp_NH_SH_mon_old-fn_3x3.csv') # made with GPT - see pamip_efp_maths

# Calculate difference
new = pd.read_csv(new_file, index_col=0)
old = pd.read_csv(old_file, index_col=0)
diff = old - new

# Calculate positive percentage difference and round to 1 decimal place
percent_diff = (abs(diff) / abs(old)) * 100
percent_diff = percent_diff.round(1)

# Rename columns to indicate percentage difference
percent_diff.columns = ['NH_pct_diff (DJF)', 'SH_pct_diff (JAS)']
# Combine absolute difference and percentage difference
combined = pd.concat([diff, percent_diff], axis=1)
# Display result
combined
```

# Previous comparison data

This is the data used in my previous comparison of regridded vs non-regridded EFP
- See this [notebook](/home/links/ct715/eddy_feedback/chapter1/resolution/pamip_efp_resolution.ipynb) for context
- ^^^ This notebook does remove the surface and the poles in the comparison. Worth bearing in mind.

```python
# Load the CSV
data_path = '/home/links/ct715/eddy_feedback/reproduce/data/efp'
file_path = os.path.join(data_path, 'previous_comparison_efp.csv')
df = pd.read_csv(file_path)

# Calculate percentage difference: ((Regridded - New) / Regridded) * 100
# Ensure positive percentages and round to 1 decimal place

df['NH_pct_diff'] = (abs(df['Regridded NH EFP'] - df['New NH EFP']) / abs(df['Regridded NH EFP'])) * 100
df['NH_pct_diff'] = df['NH_pct_diff'].round(1)

df['SH_pct_diff'] = (abs(df['Regridded SH EFP'] - df['New SH EFP']) / abs(df['Regridded SH EFP'])) * 100
df['SH_pct_diff'] = df['SH_pct_diff'].round(1)

# Handle potential division by zero (if any regridded values are zero)
df[['NH_pct_diff', 'SH_pct_diff']] = df[['NH_pct_diff', 'SH_pct_diff']].replace([np.inf, -np.inf], np.nan)

# View the updated DataFrame
df
```
