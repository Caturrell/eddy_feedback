```python
import xarray as xr
import numpy as np
import glob
```

## EC-Earth3

```python
path = '/gws/nopw/j04/arctic_connect/cturrell/CMIP6/piControl/efp_data_sit/100y/daily_averages/EC-Earth3'
files = glob.glob(f'{path}/*.nc')

# Create the time coder with cftime enabled
time_coder = xr.coders.CFDatetimeCoder(use_cftime=True)

# Use it in decode_times parameter
ds = xr.open_mfdataset(files, combine='by_coords', decode_times=time_coder)
ds
```
