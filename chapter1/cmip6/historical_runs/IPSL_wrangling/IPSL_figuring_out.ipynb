```python
files = ['/badc/cmip6/data/CMIP6/CMIP/IPSL/IPSL-CM6A-LR/historical/r1i1p1f1/6hrPlevPt/ua/gr/latest/ua_6hrPlevPt_IPSL-CM6A-LR_historical_r1i1p1f1_gr_185001010600-190001010000.nc', 
         '/badc/cmip6/data/CMIP6/CMIP/IPSL/IPSL-CM6A-LR/historical/r1i1p1f1/6hrPlevPt/va/gr/latest/va_6hrPlevPt_IPSL-CM6A-LR_historical_r1i1p1f1_gr_185001010600-190001010000.nc']

ds_ua = xr.open_dataset(files[0])
ds_va = xr.open_dataset(files[1])
```
